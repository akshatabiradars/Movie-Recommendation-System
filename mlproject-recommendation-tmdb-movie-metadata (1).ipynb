```python
import os
import ast
import pickle
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.metrics.pairwise import cosine_similarity

```

```python
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies = movies.merge(credits,on='title')
movies.head() 
```

 # select imp features

```python
movies.columns
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head() 
```

 # Extract names values from dictionary of genra and keywords columns

```python
 movies.dropna(inplace=True)
```

```python
 movies.drop_duplicates (inplace=True)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L    
    
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head() 
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

# Dictionary  Cast columns

```python
movies['cast']=movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

# Extract name values from dictionaries of director

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i ['job'] == 'Director': 
            L.append(i['name'])
    return L   
    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

 # convert overview columns string to list

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

# remove extra spaces

```python
def collapse(L):
    L1 = []
    for i in L: 
        L1.append(i.replace(" ", " "))
    return L1   
    
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse) 
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
movies.head()
```

 # combine all list into single column named tag

```python
movies['tags'] = movies['overview']
+movies['genres'] + movies['keywords']
+movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

# Build count vectorizer model

```python
cv = CountVectorizer(max_features=5000,
                     stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
vector.shape  
```

# Calculate cosine similarity

```python
similarity = cosine_similarity(vector)
similarity
```

```python
#new[new['title'] == ['The Lego  ']]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),
                      reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
    
```

```python
recommend('Gandhi')
```
